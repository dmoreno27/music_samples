```python
genius = lyricsgenius.Genius(genius_api_key_id)
artist = genius.search_artist("Beyonce", max_songs="", sort="title")
```

```python
def samples(self):
    return self._body.get('song_relationships')
```

```python
artist_samples=[]
for y in range(len(artist.songs)):
    songs_sampled=[]
    print('---------'+ artist.songs[y].title+' samples ----------')
    for x in range(len(artist.songs[y]._body.get('song_relationships')[0]['songs'])):
        songs_sampled.append({artist.songs[y]._body.get('song_relationships')[0]['songs'][x]['primary_artist']['name']:artist.songs[y]._body.get('song_relationships')[0]['songs'][x]['title']})
    artist_samples.append({artist.songs[y].title:songs_sampled})
    print(songs_sampled)
    print('-------------------')
```

```python
sampledin_list=[]
for x in range(len(artist.songs)):
    sampled_in=[]
    if len(artist.songs[x]._body.get('song_relationships')[1]['songs']) > 0:
        for y in range(len(artist.songs[x]._body.get('song_relationships')[1]['songs'])):
    #         pprint(artist.songs[x]._body.get('song_relationships')[1]['songs'][y]['primary_artist']['name'])
            sampled_in.append({(artist.songs[x]._body.get('song_relationships')[1]['songs'][y]['primary_artist']['name']):(artist.songs[x]._body.get('song_relationships')[1]['songs'][y]['title'])})
    else:
        sampled_in.append([])
    sampledin_list.append(sampled_in)
```

```python
df=pd.DataFrame()
df_songs=[]
df_artist=[]
df_writers=[]
df_writers=[]
df_year=[]
df_samples=[]
df_album=[]
df_samples_per_song=[]
df_featured_artists=[]
df_lyrics=[]
df_img=[]
a=[]
df_sampled_in=[]
df_api=[]
df_path=[]
df_description=[]
df_id=[]
df_stats=[]
df_pageviews=[]
df_apple_music_player_url=[]
df_recording_location=[]
df_release_date=[]
# annotation_count
# api_path
# apple_music_id
# apple_music_player_url
# embed_content
# featured_video
# full_title
# header_image_thumbnail_url
# header_image_url
# lyrics_owner_id
# lyrics_state
# path
# pyongs_count
# recording_location
# release_date
# song_art_image_thumbnail_url
# song_art_image_url
# stats
# title
# title_with_featured
# url
# current_user_metadata
# album
# custom_performances
# description_annotation
# featured_artists
# lyrics_marked_complete_by
# media
# primary_artist
# producer_artists
# song_relationships
# verified_annotations_by
# verified_contributors
# verified_lyrics_by
# writer_artists
# lyrics


df_writers_per_song=[]
for x in range(len(artist._songs)):
    df_songs.append(artist._songs[x].title)
    df_artist.append(artist._songs[x].artist)
    df_writers_per_song=[]
    for y in range(len(artist._songs[x].writer_artists)):
        df_writers_per_song.append(artist._songs[x].writer_artists[y]['name'])
    df_writers.append(df_writers_per_song)
    df_samples.append(list(artist_samples[x].values())[0])
    df_album.append(artist.songs[x].album)
    df_year.append(artist.songs[x].year)
    df_img.append(artist.songs[x].song_art_image_url)

    featured_artists=[]
    featured_list=[]
    y=0
    featured_list=[]
    if artist._songs[x].featured_artists == a:
        featured_artists.append([])
    else:
        for y in range(len(artist._songs[x].featured_artists)):
            featured_list.append(artist._songs[x].featured_artists[y]['name'])
    df_featured_artists.append(featured_list)
    lyrics=artist.songs[x].to_dict
    df_lyrics.append(lyrics()['lyrics'])
    
    sampled_in=[]
    if len(artist.songs[x]._body.get('song_relationships')[1]['songs']) > 0:
        for y in range(len(artist.songs[x]._body.get('song_relationships')[1]['songs'])):
            sampled_in.append({(artist.songs[x]._body.get('song_relationships')[1]['songs'][y]['primary_artist']['name']):(artist.songs[x]._body.get('song_relationships')[1]['songs'][y]['title'])})
    else:
        sampled_in.append([])
    df_sampled_in.append(sampled_in)
    df_api.append(artist._songs[x]._body['api_path'])
    df_path.append(artist._songs[x]._body['url'])
    df_id.append(artist._songs[x]._body['id'])
    df_description.append(artist._songs[x]._body['description'])
    df_apple_music_player_url.append(artist._songs[x]._body['apple_music_player_url'])
    df_recording_location.append(artist._songs[x]._body['recording_location'])
    df_release_date.append(artist._songs[x]._body['release_date'])
    df_stats.append(artist._songs[x]._body['stats'])
    df_pageviews.append(artist._songs[x]._body['pageviews'])
    


df['Performer']=df_artist
df['Song']=df_songs
df['Album']=df_album
df['Year']=df_year
df['Featuring']=df_featured_artists
df['Writers']=df_writers
df['Samples']=df_samples
df['Sampled_In']=df_sampled_in
df['Lyrics']=df_lyrics
df['Image_URL']=df_img
df['Description']=df_description
df['API_path']=df_api
df['URL']=df_path
df['Release_Date']=df_release_date

df['Music_Player_URL']=df_apple_music_player_url
df['Song_Id']=df_id
df["Reocrding_Location"]=df_recording_location
df['Stats']=df_stats
df['Pageviews']=df_pageviews

```
