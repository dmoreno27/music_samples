```python
# all_lists_by_year={}
driver = webdriver.Chrome()
driver.implicitly_wait(30)
total_links=[]
for item in clean_urls:
    url=item
    driver.get(url)
    soup=BeautifulSoup(driver.page_source, 'html.parser')
    list_temp=soup.find_all('a', href=re.compile("^.charts.(.*)"))
    results=[]
    for item in list_temp:
        result = re.search('<a href="(.*)">', str(item))
        total_links.append("https://www.billboard.com"+result.group(1))
        
```

```python
final_links=[]
for i in total_links:
    j="https://www.billboard.com"+i
    final_links.append(j)
```

```python
# final_links
```

```python

genre=[]
date=[]
df=pd.DataFrame()
for item in range(len(final_links)):
    genre.append(str(final_links[item])[33:-11])
    result=re.search('https://www.billboard.com/charts/(.*)/(.*)', str(final_links[item])).group(2)
    date.append(result)
df['Date']=date
df['Genre']=genre
df['Genre']=df['Genre'].str.replace('-',' ').str.title().str.replace('R And B','R&B').str.replace('R B ','R&B ')
df['Links']=final_links
# df
df.to_pickle("all_billboard_charts")
df_1=df.loc[df['Genre'] == 'Hot 100']
df_1.to_pickle("Hot 100")

df_2=df.loc[df['Genre'] == 'Billboard 200']
df_2.to_pickle("Billboard 200")
```
