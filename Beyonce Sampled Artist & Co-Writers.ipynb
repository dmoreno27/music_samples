```python
######## FUNCTION TO PULL DATA FOR ANY ARTIST ######## 

def sample_data(artist_func):
    artist_1 = genius.search_artist(artist_func, max_songs="", sort="popularity")
    titles_func=[]
    x=0
    for item in artist_1.songs:
        titles.append(artist_1.songs[x].title)
        x+=1
        
    writers_func=[]
    for y in range(len(artist_1._songs)):
        for x in range(len(artist_1._songs[y].writer_artists)):
            writers_func.append(artist_1._songs[y].writer_artists[x]['name'])
    
    artist_samples_func=[]
    for y in range(len(artist_1.songs)):
        songs_sampled_func=[]
        for x in range(len(artist_1.songs[y]._body.get('song_relationships')[0]['songs'])):
            songs_sampled_func.append({artist_1.songs[y]._body.get('song_relationships')[0]['songs'][x]['primary_artist']['name']:artist_1.songs[y]._body.get('song_relationships')[0]['songs'][x]['title']})
        artist_samples_func.append({artist_1.songs[y].title:songs_sampled_func})


    sampled_songs_func=[]
    a=[]
    for item in artist_samples_func:
        for key, value in item.items():
            if value != a:
                for item in value:
                    for item in list(item.values()):
                        sampled_songs_func.append(item)
                        
    sampled_artists_func=[]
    a=[]
    for item in artist_samples_func:
        for key, value in item.items():
            if value != a:
                for items in value:
                    sampled_artists_func.append(list(items.keys())[0])
    
    return Counter(sampled_artists_func)


```

```python
df=pd.DataFrame()
df_songs=[]
df_artist=[]
df_writers=[]
df_writers=[]
df_year=[]
df_samples=[]
df_album=[]
df_samples_per_song=[]
df_featured_artists=[]
df_lyrics=[]
df_img=[]
a=[]
df_sampled_in=[]

df_writers_per_song=[]
for x in range(len(artist._songs)):
    df_songs.append(artist._songs[x].title)
    df_artist.append(artist._songs[x].artist)
    df_writers_per_song=[]
    for y in range(len(artist._songs[x].writer_artists)):
        df_writers_per_song.append(artist._songs[x].writer_artists[y]['name'])
    df_writers.append(df_writers_per_song)
    df_samples.append(list(artist_samples[x].values())[0])
    df_album.append(artist.songs[x].album)
    df_year.append(artist.songs[x].year)
    df_img.append(artist.songs[x].song_art_image_url)

    featured_artists=[]
    featured_list=[]
    y=0
    featured_list=[]
    if artist._songs[x].featured_artists == a:
        featured_artists.append([])
    else:
        for y in range(len(artist._songs[x].featured_artists)):
            featured_list.append(artist._songs[x].featured_artists[y]['name'])
    df_featured_artists.append(featured_list)
    lyrics=artist.songs[x].to_dict
    df_lyrics.append(lyrics()['lyrics'])
    
    sampled_in=[]
    if len(artist.songs[x]._body.get('song_relationships')[1]['songs']) > 0:
        for y in range(len(artist.songs[x]._body.get('song_relationships')[1]['songs'])):
            sampled_in.append({(artist.songs[x]._body.get('song_relationships')[1]['songs'][y]['primary_artist']['name']):(artist.songs[x]._body.get('song_relationships')[1]['songs'][y]['title'])})
    else:
        sampled_in.append([])
    df_sampled_in.append(sampled_in)


df['Performer']=df_artist
df['Song']=df_songs
df['Album']=df_album
df['Year']=df_year
df['Featuring']=df_featured_artists
df['Writers']=df_writers
df['Samples']=df_samples
df['Sampled_In']=df_sampled_in
df['Lyrics']=df_lyrics
df['Image_URL']=df_img
```
